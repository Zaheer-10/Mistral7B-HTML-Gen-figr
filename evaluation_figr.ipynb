```python
def evaluate_model(model, dataset, tokenizer, metric_names=["rouge1", "rouge2", "rougeL", "perplexity"]):
    model.eval()
    predictions = []
    references = []

    for batch in dataset:
        inputs = tokenizer(batch["html"], return_tensors="pt", padding=True)
        with torch.no_grad():
            outputs = model.generate(**inputs, max_length=500)
        generated_texts = tokenizer.batch_decode(outputs, skip_special_tokens=True)
        predictions.extend(generated_texts)
        references.extend(batch["html"])

    scorer = rouge_scorer.RougeScorer(**{"rouge-n": {"n": 1}, "rouge-l": {}, "rouge-w": {}})
    scores = scorer.score(predictions, references)
    rouge_scores = {f"rouge-{key}": scores[key] for key in scores}

    # Calculate perplexity
    loss = model(**inputs, labels=inputs["input_ids"]).loss
    perplexity = torch.exp(loss)

    results = {**rouge_scores, "perplexity": perplexity.item()}
    return {metric_name: results[metric_name] for metric_name in metric_names}
```
